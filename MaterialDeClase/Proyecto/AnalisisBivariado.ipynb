```python
import pandas as pd
from statsmodels.stats.anova import AnovaRM
```

#ANOVARM Europa

Hipótesis nula:

No hay diferencias significativas en el uso de energía nuclear entre países de Europa desde 1965 hasta 2021.

Hipótesis alternativas:

Hay diferencias significativas en el uso de energía nuclear entre países de Europa desde 1965 hasta 2021.

-Regla de aceptación si  el valor  de p.value es mayor al 0.05  la hipótesis  nula sera aceptada.

```python
aov = AnovaRM(Europe, 'Nuclear_Electricity', 'Entity', within=['Year'])
res = aov.fit()
print(res.summary())
```

Decisión: la hipótesis nula  se rechaza ya que  el  valor  de  p.value es menor al 0.05.

Conclusión:Hay diferencias significativas en el uso de energía nuclear entre países de Europa a lo largo de varios años.
