# Linear Mixed Effects Models

### Preguntas de investigaciÃ³n que planeamos resolver con este mÃ©todo

- Â¿Existe una tendencia significativa en la generaciÃ³n de energÃ­a ni a lo largo del tiempo por continente?
- Â¿Existe una tendencia significativa en las emisiones anuales de CO2 a lo largo del tiempo por continente?


```python
dfContinents = pd.read_csv('./continents-according-to-our-world-in-data.csv',encoding='utf-8') 
dfContinents.info()
```

```python
dict_countries = dfContinents[['Entity','Continent']].set_index('Entity')
dict_countries
```

```python
dict_countries = dict_countries['Continent'].to_dict()
dict_countries
```

```python
df['Continent'] = df['Entity'].map(dict_countries,na_action='ignore')
df[['Continent','Entity']]
```

```python
import statsmodels.formula.api as smf
md = smf.mixedlm("Nuclear_Electricity ~ Year", df, groups=df["Continent"]) 
mdf = md.fit(method=["lbfgs"]) 
print(mdf.summary())
```

# Conclusiones a partir de los datos otorgados por el modelo
- El modelo indica que la producciÃ³n de electricidad nuclear no presenta una intercepciÃ³n significativa en el aÃ±o cero, lo que podrÃ­a indicar que la producciÃ³n de electricidad nuclear en ese momento era cercana a cero.

- El coeficiente de aÃ±o indica que, en promedio, la producciÃ³n de electricidad nuclear aumentÃ³ en 1.481 unidades por aÃ±o (esto es significativo, con un valor p < 0.05).

- El modelo tambiÃ©n indica que la varianza de grupo es cero, lo que significa que no hay variaciÃ³n en la producciÃ³n de electricidad nuclear entre los diferentes continentes.

```python

md = smf.mixedlm("Annual_CO2_emissions ~ Year", df, groups=df["Continent"]) 
mdf = md.fit(method=["lbfgs"]) 
print(mdf.summary())
```

# Conclusiones a partir de los datos otorgados por el modelo

- El coeficiente del aÃ±o (Year) es -0.018, lo que significa que, en promedio, las emisiones de diÃ³xido de carbono disminuyen en 0.018 toneladas por aÃ±o. 

- El modelo tambiÃ©n indica que la varianza de grupo es cero, o que indica que no hay diferencias significativas en las emisiones de diÃ³xido de carbono entre los cinco continentes.
