```python
import pandas as pd
from statsmodels.stats.anova import AnovaRM
```

#ANOVARM Europa

HipÃ³tesis nula:

No hay diferencias significativas en el uso de energÃ­a nuclear entre paÃ­ses de Europa desde 1965 hasta 2021.

HipÃ³tesis alternativas:

Hay diferencias significativas en el uso de energÃ­a nuclear entre paÃ­ses de Europa desde 1965 hasta 2021.

-Regla de aceptaciÃ³n si  el valor  de p.value es mayor al 0.05  la hipÃ³tesis  nula sera aceptada.

```python
aov = AnovaRM(Europe, 'Nuclear_Electricity', 'Entity', within=['Year'])
res = aov.fit()
print(res.summary())
```

DecisiÃ³n: la hipÃ³tesis nula  se rechaza ya que  el  valor  de  p.value es menor al 0.05.

ConclusiÃ³n:Hay diferencias significativas en el uso de energÃ­a nuclear entre paÃ­ses de Europa a lo largo de varios aÃ±os.
